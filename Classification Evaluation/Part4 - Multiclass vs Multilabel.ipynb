## Multinomial

Sebelumnya kita melakukan klasifikasi menggunakan regresi, yaitu dengan menggunakan fungsi sigmoid. Multinomial menggunakan fungsi sigmoid yang lebih umum, yaitu **fungsi softmax**. Oleh sebab itu, **Multinomial Logistic Regression** sering juga disebut sebagai **Softmax Regression**

## **Softmax (Optional)**

```python
import numpy as np
import pandas as pd
```

```python
state = np.random.RandomState(42)
df = pd.DataFrame(state.randint(0, 10, 5), columns=['output'])
df['exp'] = df.output.apply(np.exp)
df['softmax'] = df.exp.transform(lambda x: x/x.sum())
df['log_softmax'] = df.exp.transform(lambda x: np.log(x/x.sum()))
```
