More Simple Model -> Low Variance, but High bias -> Prone to Underfit

## Exerice - Random Forest Parameter
- increase n_estimators => more averaging => decrease overfit
- increase max_depth => more decision => low bias => prone to overfit
- increase min_sample_split => prevent splitting => less depth => decrease overfit
- increase max_features => prone to get best decision => prone to overfit
- increase min_impurity_decrease => prevent splitting => less depth => decrease overfit
- increase min_samples_leaf => prevent splitting => less depth => decrease overfit
